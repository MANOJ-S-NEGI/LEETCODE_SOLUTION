#include <iostream>
#include <vector>
using namespace std;

void rotate(vector<int>& nums, int k)
{
    int n = nums.size();
    k = k % n;

    // Reverse entire array
    for (int i = 0, j = n - 1; i < j; i++, j--)
        swap(nums[i], nums[j]);

    // Reverse first k elements
    for (int i = 0, j = k - 1; i < j; i++, j--)
        swap(nums[i], nums[j]);

    // Reverse remaining elements
    for (int i = k, j = n - 1; i < j; i++, j--)
        swap(nums[i], nums[j]);
}

int main()
{
    vector<int> nums = {3, 4, 5, 6, 7, 8, 9};
    int k = 4;
    rotate(nums, k);

    cout << "[ ";
    for (int num : nums)
    {
        cout << num << " ";
    }
    cout << "]" << endl;

    return 0;
}

```
#include <iostream>
#include <vector>
using namespace std;

int maxProfit(vector<int>& prices)
 {
    if (prices.empty()) return 0;

    int min_price = prices[0];
    int max_profit = 0;

    for (int i = 1; i < prices.size(); i++)
    {
        if (prices[i] < min_price)
        {
            min_price = prices[i];
        }
        else
        {
            int profit = prices[i] - min_price;
            if (profit > max_profit)
            {
                max_profit = profit;
            }
        }
    }
    return max_profit;
}

int main() {
    vector<int> prices = {7,6,5,4,3,2,1};
    int res = maxProfit(prices);
    cout << "Max Profit: " << res << endl;
    return 0;
}


```

Sell and Buy Stocks:

```

#include <iostream>
#include <vector>
using namespace std;

int maxProfit(vector<int>& prices)
 {
    bool is_ascend = 1;
    bool is_descend = 1;

    int total_profit = 0;
    int current_profit = 0;

    for (int i = 1; i < prices.size(); i++)
    {
        if (is_ascend)
        {
            is_ascend = (prices[i] > prices[i-1])?true:false;
        }

        if (is_descend)
        {
            is_descend = (prices[i] < prices[i-1])?true:false;
        }

        if(prices[i] >= prices[i-1])  //case 1 {7,1,5,3,6,4};  3>=6     //  case 2 [6,1,3,2,4,7]
        {
            current_profit = prices[i] - prices[i-1];    // 6-3  = 3   // 4- 2 = 2
            total_profit = total_profit + current_profit;  // 4+3 =7   // tf = 2+2
        }
    }

    if(is_descend)
    {
        return 0;
    }
    else if(is_ascend)
    {
        return prices[prices.size()-1] - prices[0];
    }
    else
    {
        return total_profit;
    }
}

int main() {
    vector<int> prices = {7,1,5,3,6,4};
    //vector<int> prices = {1,2,3,4,5,6,7,8,9};
    //vector<int> prices = {7,6,5,4,3,2,1};

    int res = maxProfit(prices);
    cout << "Max Profit: " << res << endl;
    return 0;
}



```

```
#include <iostream>
#include <vector>
using namespace std;

bool canJump(vector<int>& nums)
{
    int index_reached = 0;
    int n = nums.size();

    for (int i = 0; i < n; i++)
    {
        if (i > index_reached)
            return false;

        int current_Reach = i + nums[i];

        if (current_Reach > index_reached)
            index_reached = current_Reach;

        if (index_reached >= n - 1)
            return true;
    }

    return 0;
}

int main() {
    vector<int> jump = {3, 4, 1, 2, 3};
    bool res = canJump(jump);
    cout << "Can Jump: " << (res ? "true" : "false") << endl;
    return 0;
}


```

```

#include <iostream>
#include <vector>
using namespace std;

int jump(vector<int>& nums)
{
    int n = nums.size();
    int jumps = 0, farthest = 0, current_end = 0;

    for (int i = 0; i < n - 1; i++)
    {
        farthest = max(farthest, i + nums[i]);  // Update max reach

        if (i == current_end)
        {
            jumps++;
            current_end = farthest;

            if (current_end >= n - 1)
                break;
        }
    }
    return jumps;
}

int main() {
    vector<int> nums = {2, 3, 1, 1, 4};
    cout << "Minimum jumps: " << jump(nums) << endl;
    return 0;
}


```

Gas Station Problem:


```
#include<iostream>
#include<vector>
using namespace std;

class Solution
{
public:
    int canCompleteCircuit(vector<int>& gas, vector<int>& cost)
    {
        int total_gas = 0, total_cost = 0, tank = 0, index = 0;
        int start_station;
        for(int i = 0; i<gas.size(); i++)
        {
            total_gas = total_gas + gas[i];
            total_cost = total_cost + cost[i];
        }

        if(total_cost > total_gas)return -1;
        for(int i = 0; i<gas.size(); i++)
        {
            tank = tank + gas[i] - cost[i];
            if(tank < 0)
            {
                tank = 0;
                index = i + 1;
            }
        }
        return index;
    }
};

int main()
{
    Solution obj;
    vector<int>gas = {5, 1, 2, 3, 4};
    vector<int>cost= {4, 4, 1, 5, 1};

    cout<<obj.canCompleteCircuit(gas, cost);
    return 0;
}

```

CANDY PROBLEM

```

#include <iostream>
#include <vector>
#include <algorithm>
using namespace std;

class Solution {
public:
    int candy(vector<int>& ratings)
    {
        int n = ratings.size();
        if(n == 0) return 0;
        vector<int> candies(n, 1);

        for (int i = 1; i < n; i++)
        {
            if (ratings[i] > ratings[i - 1])
            {
                candies[i] = candies[i - 1] + 1;
            }
        }


        for (int i = n - 2; i >= 0; i--)
        {
            if (ratings[i] > ratings[i + 1])
            {
                candies[i] = max(candies[i], candies[i + 1] + 1);

            }
        }

        int total = 0;
        for (int c : candies)
        {
            total += c;
        }

        return total;
    }
};

int main() {
    Solution obj;
    vector<int> ratings = {1, 2, 3, 2, 2, 3};
    cout << "Total Candies Required: " << obj.candy(ratings) << endl;
    return 0;
}
```

Trapped Water:

```
class Solution {
public:
    int trap(vector<int>& height) {
        int n = height.size();
        if(n == 0) return 0;

        // Pre compute left maximums
        vector<int> left_max(n, 0);
        left_max[0] = height[0];
        for (int i = 1; i < n; i++) {
            left_max[i] = max(left_max[i-1], height[i]);
        }

        // Pre compute right maximums
        vector<int> right_max(n, 0);
        right_max[n-1] = height[n-1];
        for (int i = n - 2; i >= 0; i--) {
            right_max[i] = max(right_max[i+1], height[i]);
        }

        // Calculate trapped water
        int water = 0;
        for (int i = 0; i < n; i++) {
            water += min(left_max[i], right_max[i]) - height[i];
        }

        return water;
    }
};


```

```
#include <iostream>
using namespace std;

class Solution
{
public:
    string reverseWords(string s)
    {
        string result = "";
        string word = "";

        for (int i = 0; i < s.size(); i++)
        {
            if (s[i] != ' ')
            {
                word += s[i];
            }
            else
            {
                if (!word.empty())
                {
                    if (!result.empty())
                        result = " " + result;

                    result = word + result;
                    word = ""; // Reset for next word
                }
            }
        }

        if (!word.empty())
        {
            if (!result.empty())
                result = " " + result;
            result = word + result;
        }
        return result;
    }
};

int main()
{
    Solution obj;
    string input = " hello world ";
    cout << obj.reverseWords(input) << endl;
    return 0;
}


```
